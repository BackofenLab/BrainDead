```python
import pandas as pd
df_hekblue_x = pd.read_excel('../data/activation-data/processed-HEKBlue.xlsx', index=True, index_col=0)
df_hekblue = df_hekblue_x.transpose().reset_index(drop=True)
df_hekblue['rep1'] = pd.to_numeric(df_hekblue['rep1'])
df_hekblue['rep2'] = pd.to_numeric(df_hekblue['rep2'])
df_hekblue['rep3'] = pd.to_numeric(df_hekblue['rep3'])
df_hekblue['rep4'] = pd.to_numeric(df_hekblue['rep4'])
df_hekblue['mean'] = df_hekblue.mean(axis=1, numeric_only=True)
df_hekblue['std'] = df_hekblue.std(axis=1, numeric_only=True)
df_hekblue
```

```python
df_tnf_x = pd.read_excel('../data/activation-data/processed-TNF.xlsx', index=True, index_col=0)
df_tnf = df_tnf_x.reset_index(drop=True)
df_tnf['n=1'] = pd.to_numeric(df_tnf['n=1'],errors='coerce')
df_tnf['n=2'] = pd.to_numeric(df_tnf['n=2'],errors='coerce')
df_tnf['n=4'] = pd.to_numeric(df_tnf['n=4'],errors='coerce')
df_tnf['n=5'] = pd.to_numeric(df_tnf['n=5'],errors='coerce')
df_tnf['mean'] = df_tnf.mean(axis=1, numeric_only=True)
df_tnf['std'] = df_tnf.std(axis=1, numeric_only=True)
df_tnf
```

```python
df_merged = pd.merge(df_hekblue, df_tnf, left_on=['List-type','miRNA'],suffixes=['_HEK','_TNF'], right_on=['List-type','miRNA'])

df_miRNAs = pd.read_excel('../data/activation-data/miRNAs.xlsx')

df_merged_miRNAs = pd.merge(df_miRNAs, df_merged, left_on='miRNA', right_on='miRNA')  
df_merged_miRNAs
```

```python
df_merged_miRNAs['miRNA-label'] = df_merged_miRNAs['miRNA']
df_merged_miRNAs.loc[df_merged_miRNAs['Predicted_class']==0,'miRNA-label'] = ''
df_merged_miRNAs.replace({'Predicted_class':0},'non-activating', inplace=True)
df_merged_miRNAs.replace({'Predicted_class':1},'activating', inplace=True)
df_merged_miRNAs
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
from adjustText import adjust_text

def label_point(a, x, y, label, hue, style):
    sns.set_style('white')
    fig = plt.figure(figsize=(4.5,4.5))
    ax = fig.add_subplot(111)
    table(ax,a[a['Predicted_class']=='activating'][['rank','miRNA']].set_index('miRNA'), 
          loc='right',bbox=[1.4, .0, .12, .9],
          cellLoc='center',rowLoc='center',colLoc='center',
         #colWidths=[0.1, 0.01, 0.1]
         )

    sns.scatterplot(data=a, x=x, y=y,ax=ax, hue=hue, palette=['#0908c5','#f17f22'],
                    style=style, markers=['o', 's', '+'], s=50, alpha=0.7, edgecolor='k'
                   )
    ax.set_ylabel(ax.get_ylabel().replace('mean_TNF',r'TNF-$\alpha$ (pg/mL)'))
    ax.set_xlabel(ax.get_xlabel().replace('mean_HEK','OD SEAP\nFold change'))

    texts = []

    for i, point in a.iterrows():
        if point['Predicted_class']=='activating':
            texts.append(ax.text(point[x], point[y], str(point[label])))
    
    adjust_text(texts, expand_text=(1.5,1.5), expand_objects=(1.5,1.5),expand_points=(2,2),
                #only_move={'points':'y', 'texts':'y'}, 
                arrowprops=dict(arrowstyle="-", color='gray', lw=0.5))
    fig.savefig('./scatter_activation.svg', bbox_inches='tight')
    fig.savefig('./scatter_activation.png', dpi=500, bbox_inches='tight')

        
from pandas.plotting import table 

# $ax.errorbar(x=df_merged['mean_TNF'], y=df_merged['mean_HEK'], xerr=df_merged['std_TNF'], yerr=df_merged['std_HEK'],fmt='o')
# label_point(df_merged_miRNAs, 'mean_TNF', 'mean_HEK', 'miRNA', ax)
# label_point(df_merged_miRNAs, 'mean_HEK', 'prediction_score', 'miRNA', 'class')
# label_point(df_merged_miRNAs, 'mean_TNF', 'prediction_score', 'miRNA', 'class')
# label_point(df_merged_miRNAs, 'mean_TNF', 'mean_HEK', 'miRNA-label', 'class','List-type')
# label_point(df_merged_miRNAs, 'mean_TNF', 'mean_HEK', 'rank', 'class','List-type')
label_point(df_merged_miRNAs,  'mean_HEK', 'mean_TNF','rank', 'Predicted_class','List-type')
```
